### Activation Functions

The following cell contains the functions for all the activation functions that will be used in this experiment as well as functions for their derivatives.
The activations being used are:
- Rectified Linear Unit (ReLU)
- Sigmoid
- linear
- tanh

```python
# Activation functions

def ReLU(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def linear(x):
    return x

def tanh(x):
    return 2/(1 + np.exp(-2*x)) - 1

def dif_ReLU(x):
    if(x >= 0):
        return 1
    else:
        return 0
    
def dif_sigmoid(x):
    return sigmoid(x)*(1 - sigmoid(x))

def dif_linear(x):
    return 1

def dif_tanh(x):
    return 1/(np.square(x) + 1)

       
    
```

## Structure of the Neural Network

The implementation of the neural network is found in the cell below. There are two functions, one that creates and trains the neural network with backpropagation and gradient descent called feed_forward(); and one function that predicts the output classification by passing the x variables through the neuraal network with the learned weights called predict().

The structure of the neural network contains 23 nodes in the input layer - one for each input variable in the dataset. One hidden layer in whoch the amount of nodes is one of the hyperparameters that I tune to achieve the best performance. There is one output node as this is a binary classification problem and therefore only need an output of 0 or 1.

The there are two weight matricies, w1 is the matrix from the input layer to the hidden layer and has the dimensions [no. of hidden nodes x 23]. w2 is the weight matrix from the hidden layer to the output layer and has the dimensions [1 x no. of hidden nodes]. 

Each training sample is passed through the network and a prediction is made. The difference between the prediction and the classification label of the sample is calculated and backpropagated back through the network in order to adjust the wights of the network. We calculate the mean squared error (MSE) of the entire training dataset and we iterate through the training dataset until the error converges. Each iteration over the whole training dataset is called an epoch. I use L2 regularization in order to assist with the generalizability of the network and so that the network does not overfit the training data.

In the predict() function, I pass the x variables of the validation/testing dataset through the neural network with the learned weights from the feed_foward() function in order to get a prediction. The predictions are stored in an array and then compared to the y variables of the respective dataset.

```python
def feed_forward(x, hidden, y, alpha, epoch, lmda, act, dif_act):
    error_total = []
    
    np.random.seed(10)
    
    w1 = np.random.randn(hidden, len(x.columns))/np.sqrt(len(x.columns))
    bias1 = np.random.randn(hidden,1)
    
    w2 = np.random.randn(1, hidden)/np.sqrt(hidden)
    bias2 = np.random.randn(1,1)
    mse_old = 100000
    diff = 100000
    ind = 0
    while(diff > 0.00001 and ind < epoch):
        z_arr = []

        for i in range(len(x)):

            a1 = x.iloc[i].values.reshape((len(x.columns),1))
            #print(a1)
            z2 = np.dot(w1, a1) + bias1
            a2 = act(z2)

            z3 = np.dot(w2, a2) + bias2
            a3 = sigmoid(z3)


            z_arr.append(float(a3))

            yhat = a3
            #print(yhat)
            ### BACKPROPAGATION ###

            d3 = yhat - y.iloc[i]
            d2 = np.dot(w2.T, d3)*dif_sigmoid(z3)

            J1 = np.dot(d2,a1.reshape((1,len(x.columns)))) + lmda*np.sum(np.square(w1))  #L1 = np.sum(np.abs(w1)), L2 = np.sum(np.square(w1))
            J2 = np.dot(d3,a2.reshape((1,hidden))) + lmda*np.sum(np.square(w2))
            
            w1 -= alpha*J1 
            w2 -= alpha*J2 
            
        yhat_total = np.array(z_arr)
        
        mse = 1/len(x)*np.sum(np.square(yhat_total - y))
        print(mse)
        acc = (1-mse)
        diff = mse_old - mse
        mse_old = mse
        #print(diff)
        ind +=1
        #print(ind)
        error_total.append(mse)
        
        
        
    return error_total, w1, w2
    
def predict(x, hidden, w1, w2, act, boundary):
    np.random.seed(10)
    bias1 = np.random.randn(hidden,1)
    bias2 = np.random.randn(1,1)
    predictions = []
    for i in range(len(x)): #len(x)

            a1 = x.iloc[i].values.reshape((len(x.columns),1))
            z2 = np.dot(w1, a1) + bias1
            a2 = act(z2)

            z3 = np.dot(w2, a2) + bias2
            a3 = sigmoid(z3)

            yhat = a3
            #print(yhat)
            if(yhat < boundary):
                predictions.append(0)
            else:
                predictions.append(1)
                
    return predictions
```

The training, testing and validation datasets are split into x and y variables in order to effectively train and predict the classification.

```python
x_train = df_train_new.drop(["Risk"], axis = 1)
y_train = df_train_new["Risk"]

x_test = df_test.drop(["Risk"], axis = 1)
y_test = df_test["Risk"].reset_index(drop=True)

x_validate = df_validate.drop(["Risk"], axis = 1)
y_validate = df_validate["Risk"].reset_index(drop=True)
```

## Implementation of the neural network
In this section, I train the neural network on the training data with different activation functions to test the performance.
I also tune the hyperparameters on the model to achieve the highest prediction possible before predicting on the testing dataset.
The hyperparameters that are tuned are:

- The amount of nodes in the hidden layer - hidden
- The learning rate - alpha
- The amount of regularization - lamda

The results of the training of the model are shown through a lineplot of the MSE of each epoch over the amount of epochs.
A final training error percentage is displayed.

The results of the predictions are shown through a confusion matrix with the validation accuracy of the predictions below.

```python
#hyperparameters:
hidden = 10 #16 = best
alpha = 0.1 # 0.00005 = best
lamda = 0.0 # 0.0001 = best for L2 regularization

#activation function:
activation = sigmoid
dif_activation = dif_sigmoid
boundary = 0.5

#Train neural network:
error, theta1, theta2 = feed_forward(x_train, hidden, y_train, alpha, 400, lamda, activation, dif_activation)

#Plot error
print("-------- Results From Training --------")
plt.plot(np.arange(0,len(error)),error)
plt.title("Error over Time")
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.show()  
print("Final Training Error: ", error[-1]*100,"%")

#Predict on the validation set
predictions = predict(x_validate, hidden, theta1, theta2, activation, boundary)
pred_series = pd.Series(predictions)
print("\n")
#Plot results on confusion matrix
print("-------- Results From Predicting --------")
confusion_matrix = pd.crosstab(y_validate, pred_series, rownames=['Actual'], colnames = ['Predicted'])
sn.heatmap(confusion_matrix, annot=True,cmap="YlGnBu")
plt.show()
print("\n")
display(confusion_matrix)
print("\n")
accuracy = (confusion_matrix[0][0]+confusion_matrix[1][1])/len(x_validate)
print("Validation Accuracy: ", accuracy)
```

## Adding more layers to the neural network

In this section, I experiment with the architecture of the neural network by adding a new hidden layer to asses performance. The implementation of the training and prediction is kept the same to the initial neural network, the only difference is that a new hidden layer is added between the first hidden layer and the output layer.  

```python
def deep_feed_forward(x, hidden, hidden2, y, alpha, epoch, lmda, act, dif_act):
    error_total = []
    np.random.seed(10)
    
    w1 = np.random.randn(hidden, len(x.columns))/np.sqrt(len(x.columns))
    bias1 = np.random.randn(hidden,1)

    w2 = np.random.randn(hidden2, hidden)/np.sqrt(hidden)
    bias2 = np.random.randn(hidden2,1)

    w3 = np.random.randn(1, hidden2)/np.sqrt(hidden2)
    bias3 = np.random.randn(1,1)

    mse_old = 100000
    diff = 100000
    ind = 0
    
    while(diff > 0.00001 and ind < epoch):
        z_arr = []

        for i in range(len(x)):

            a1 = x.iloc[i].values.reshape((len(x.columns),1))
            z2 = np.dot(w1, a1) + bias1
            a2 = act(z2)

            z3 = np.dot(w2, a2) + bias2
            a3 = act(z3)
            
            z4 = np.dot(w3, a3) + bias3
            a4 = act(z4)

            z_arr.append(float(a4))

            yhat = a4

            ### BACKPROPAGATION ###

            d4 = yhat - y.iloc[i]
            d3 = np.dot(w3.T, d4)*dif_act(z3)
            d2 = np.dot(w2.T, d3)*dif_act(z2)
           
            J1 = np.dot(d2,a1.reshape((1,len(x.columns)))) + lmda*np.sum(np.square(w1))
            J2 = np.dot(d3,a2.reshape((1,hidden))) + lmda*np.sum(np.square(w2))
            J3 = np.dot(d4,a3.reshape((1,hidden2))) + lmda*np.sum(np.square(w3))

            w1 -= alpha*J1  
            w2 -= alpha*J2  
            w3 -= alpha*J3  

        yhat_total = np.array(z_arr)
        mse = 1/len(x)*np.sum(np.square(yhat_total - y))
        diff = mse_old - mse
        mse_old = mse
        ind +=1
        error_total.append(mse)
        
    return error_total, w1, w2, w3

def deep_predict(x, hidden, hidden2, w1, w2, w3, act, boundary):
    np.random.seed(10)
    bias1 = np.random.randn(hidden,1)
    bias2 = np.random.randn(hidden2,1)
    bias3 = np.random.randn(1,1)
    predictions = []
    for i in range(len(x)):

            a1 = x.iloc[i].values.reshape((len(x.columns),1))
            z2 = np.dot(w1, a1) + bias1
            a2 = act(z2)

            z3 = np.dot(w2, a2) + bias2
            a3 = act(z3)
            
            z4 = np.dot(w3, a3) + bias3
            a4 = act(z4)
            yhat = float(a4)
            
            if(yhat >= boundary):
                predictions.append(1)
            else:
                predictions.append(0)
                
    return predictions
```
